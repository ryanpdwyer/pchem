```python
def hex_pts(x=0, y=0, s=2/3, angle=0):
    angles = (np.arange(7)*60-angle)*(np.pi/180)
    x_pts = np.cos(angles)*s + x
    y_pts = np.sin(angles)*s + y
    return x_pts, y_pts


```
