```python
def lookup_molecule_info(molecule_name):
    c1 = cs.search(molecule_name)
    if len(c1) > 0:
        c1_result = c1[0] 
        molecule_info = dict(
        molecule_name = molecule_name,
        molecular_weight=c1_result.molecular_weight,
         molecular_formula=c1_result.molecular_formula,
         record_id=c1_result.record_id,
         number_of_results = len(c1))
    else:
        molecule_info = dict(
        molecule_name = molecule_name,
        molecular_weight=np.nan,
        molecular_formula="NA",
        record_id="NA",
        number_of_results = len(c1))
    
    return molecule_info
```
