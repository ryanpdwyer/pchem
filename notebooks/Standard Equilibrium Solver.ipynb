```python
import sympy as sm
from sympy.abc import *
import pchem as pc
```

# Sympy solving the standard equilibrium equation

You can do this analysis on your calculator or on (https://js.munano.org) (Sympy solver) as well. See the book Chapter 12:




```python
R = 8.3145 # J/mol-K
eqn = H-T*S + R*T*sm.log(Q) # This form assumes ΔrG = 0
# Solve for K at 330 K given ΔH°=30e3 J/mol (30 kJ/mol) and ΔS°=60 J/mol-K
pc.solve(eqn, Q, dict(H=30e3, S=60, T=330))
```

Just copy and paste the last line to do different analyses. For example, at what temperature would K be equal to 0.01?

```python
pc.solve(eqn, T, dict(H=30e3, S=60, T=330, Q=0.01))
```
