```python
import matplotlib.pyplot as plt
import numpy as np

# Constants
kB = 1.381e-23 #J/K
n = 1 # We'll use a constant 1 mol of total particles.

def nB(delta_eps, T):
    return n * np.exp(-2*delta_eps/(kB * T))       # translate the equation into Python (np.exp, etc)

nB(1e-21,300)

def nA(delta_E, T):
    return n - nB(delta_E, T)
nA(1e-21, 300)

temps = np.linspace(5, 400)

delta_E = 1.66e-21 #J
plt.plot(temps, nA(delta_E, temps), label="$n_A$")
```

```python
temps = np.linspace(5, 400)

delta_E = 1.66e-21 #J
plt.plot(temps, nA(delta_E, temps), label="$n_A$")
plt.plot(temps, nB(delta_E, temps), label="$n_B$")
plt.grid()
plt.legend()
plt.xlabel("Temperature (K)")
plt.ylabel("Particles(mols)")
```
