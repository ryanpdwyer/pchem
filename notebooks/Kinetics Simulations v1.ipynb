```python
from scipy.spatial import ConvexHull
from munch import Munch
```

```python
def prism(s, h, offset=0):
    pts = np.array([
        [0, 0, 0],
        [s/2, s*np.sqrt(3)/2, 0],
        [-s/2, s*np.sqrt(3)/2, 0],
        [0, 0, h],
        [-s/2, s*np.sqrt(3)/2, h],
        [s/2, s*np.sqrt(3)/2, h],
        ])
    x, y, z = pts.T
    z = z + offset
    ch = ConvexHull(pts)
    return Munch(x=x, y=y, z=z, triangles=ch.simplices)
    
```

```python
out = prism(5, 0.2, -0.3)
out2 = prism(2, 0.2)
```

```python
class Enzyme:
    def __init__(self, kf, kr, kcat):
        self.kf = kf
        self.kr = kr
        self.kcat = kcat
    
    def dydt(self, y, t):
        """State vector is S, E, ES, P"""
        kf, kr, kcat = self.kf, self.kr, self.kcat
        r1fwd = kf*y[0]*y[1]
        r1rev = kr * y[2]
        r2fwd = kcat * y[2]
        return np.array([
            -r1fwd + r1rev,
            -r1fwd + r1rev + r2fwd,
            r1fwd - r1rev - r2fwd,
            r2fwd
        ])
    
    def rate(self, y):
        return self.kcat * y.T[2]
```

```python
e = Enzyme(1, 1, 1)
y0 = np.array([1, 0.1, 0, 0])
t= np.linspace(0, 1000, 1001)
out = integrate.odeint(e.dydt, y0, t)
```
