```python
# Constants
kB = 1.381e-23 # J/K
n = 1 # We'll use a constant 1 mol of total particles.


def nB(delta_eps, T):
    return # translate the equation into Python (np.exp, etc)

def nB(delta_eps, T):
    return n * np.exp(-2*delta_eps/(kB*T))

nB(1e-21, 300)
```

```python
def nA(delta_E, T):
    return n - nB(delta_E, T)

nA(1e-21, 300)
```

```python


def eq_constant(delta_E, T):
    return nB(delta_E, T) / nA(delta_E, T)
```

```python
temps = np.linspace(1, 400)
K = eq_constant(1.5e-21, temps)

plt.plot(temps, K)
plt.xlabel("Temperature (K)")
```
