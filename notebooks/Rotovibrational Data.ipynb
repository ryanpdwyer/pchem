```python
class Evib:
    def __init__(self, 
                    omega_e, 
                 omega_e_x_e,
                 omega_e_y_e=0):
        self.we = omega_e
        self.wexe = omega_e_x_e
        self.weye = omega_e_y_e
    def __call__(self, i):
        return (self.we*(i+1/2) -
            self.wexe*(i+1/2)**2 + 
                self.weye*(i+1/2)**2)
    def __repr__(self):
        return f"Evib(w={self.we}, wexe={self.wexe})"

class Erot:
    def __init__(self, 
                    omega_e, 
                 omega_e_x_e,
                 Be,
                 alpha_e,
                 gamma_e,
                 De,
                 beta_e,
                 omega_e_y_e=0):
        self.we = omega_e
        self.wexe = omega_e_x_e
        self.weye = omega_e_y_e
        self.Be = Be
        self.alpha_e = alpha_e
        self.gamma_e = gamma_e
        self.De = De
        self.beta_e = beta_e

    def __call__(self, ivib, irot):
        Bv = self.Be - self.alpha_e*(ivib+1/2) + self.gamma_e*(ivib+1/2)
        Dv = self.De - self.beta_e*(ivib+1/2)
        return Bv*(irot)*(irot+1) - Dv * irot**2 * (irot+1)**2

    def __repr__(self):
        return f"Erot(w={self.we}, wexe={self.wexe})"
```

```python
clf = Evib(786.15, 6.16)
clfRot = Erot(786.15, 6.16, 0.5164805, 0.0043385, 0, 8.77e-7, 0)
```
